расходные характеристики ГЭС

```python
Q11=[]
Q22=[]
Q33=[]
for i in range (0,15,1):
  Q11.append(100*N1[i]*1000/(9.81*n[i]*H_ras))
  Q22.append(100*N2[i]*1000/(9.81*n[i]*H_ras))
  Q33.append(100*N3[i]*1000/(9.81*n[i]*H_ras))
Q11_max=[]
Q22_max=[]
Q33_max=[]
for i in range (0,11,1):
  Q11_max.append(100*N1_max[i]*1000/(9.81*n[i]*H_max))
  Q22_max.append(100*N2_max[i]*1000/(9.81*n[i]*H_max))
  Q33_max.append(100*N3_max[i]*1000/(9.81*n[i]*H_max))
Q11_min=[]
Q22_min=[]
Q33_min=[]
for i in range (0,13,1):
  Q11_min.append(100*N1_min[i]*1000/(9.81*n[i]*H_min))
  Q22_min.append(100*N2_min[i]*1000/(9.81*n[i]*H_min))
  Q33_min.append(100*N3_min[i]*1000/(9.81*n[i]*H_min))
```

```python
fig, ax = plt.subplots()
plt.ioff()
ax.plot(Q11,N1,Q22,N2,Q33,N3,Q11_max,N1_max,Q22_max,N2_max,Q33_max,N3_max,Q11_min,N1_min,Q22_min,N2_min,Q33_min,N3_min, linestyle='solid')

qw = 300
plt.axvline(x=qw, color='r', linestyle='--', label=qw)
qw1 = 580
plt.axvline(x=qw1, color='r', linestyle='--', label=qw1)

lgnd = ax.legend(['1 ГА расч', '2 ГА расч', '3 ГА расч', '1 ГА макс', '2 ГА макс', '3 ГА макс', '1 ГА мин', '2 ГА мин', '3 ГА мин'], shadow=True)
lgnd.get_frame().set_facecolor('#ffb19a')


plt.show()

```
